## Preparing the dataset

```python
dataset = pd.DataFrame(boston_df.data, columns=boston_df.feature_names)
```

```python
dataset['Price'] = boston_df.target
```

```python
dataset.head()
```

## Quick analysis

```python
dataset.info()
```

```python
dataset.describe()
```

## Exploratory data analysis

```python
# Let's check the correlation between features (multicollinearity) and against the independent feature.

import seaborn as sns
sns.set(rc={'figure.figsize':(10,6)})
cmap = sns.diverging_palette(500, 10, as_cmap=True)
sns.heatmap(dataset.corr(), cmap=cmap, annot=True)
```

```python
sns.scatterplot(x=dataset['CRIM'],y=dataset['Price'])
```

visualize crime rate with price, notably if the crime rate increase the price will drop the property.

```python
# visualize average number of rooms with price
sns.scatterplot(x=dataset['RM'],y=dataset['Price'])
```

Here, we can see clearly a positive correlation. Let's plot the regression line between the average number of rooms and the price.

```python
sns.regplot(x=dataset['RM'], y=dataset['Price'])
```

```python
sns.regplot(x=dataset['LSTAT'], y=dataset['Price'])
```

From the heatmap correlation, the lower status of the population (LSTAT) got the worst correlation with price. 

```python
sns.regplot(x=dataset['CHAS'], y=dataset['Price'])
```

As we can see here, it's hardly correlated. There is no linearity with price, we need to understand the linearity should definitely be there in the dataset. To create a better regression model.

```python
# Let's make the independent features and dependent feature.
X = dataset.iloc[:,:-1]
y = dataset.iloc[:,-1]
print(X.head())
print(y.head())
```

```python
# Let's split X,y to train and test
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# let's scale the X_train and X_test
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
X_train
```

```python
X_test
```

```python
import pickle
pickle.dump(scaler,open('scaling.pkl','wb'))
```

## Model training

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(X_train, y_train)
```

```python
# Let's print the coefficient of the independent features
print(reg.coef_)
```

These are the coefficients of the independent features. For example, the coefficient represents the Crime rate (CRIM). 

```python
print(reg.intercept_)
```

```python
print(reg.get_params())
```

## Prediction with test data

```python
reg_pred = reg.predict(X_test)
```

```python
reg_pred
```

```python
# Let's plot the prediction with y_test
sns.scatterplot(x=y_test,y=reg_pred)
```

```python
# Let's calculate the residuals and plot it
residuals = y_test - reg_pred
sns.displot(residuals, kind='kde')
```

It looks like a normal distribution, but here on the right-hand side, there are outliers and errors, and it's quite high. In general, the model performs well

```python
sns.scatterplot(x=reg_pred, y= residuals)
```

as we can see it's uniform distribution

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
print(mean_absolute_error(y_test, reg_pred))
print(mean_squared_error(y_test, reg_pred))
print(np.sqrt(mean_squared_error(y_test, reg_pred)))
```

## R square and adjusted R square

```python
from sklearn.metrics import r2_score
score = r2_score(y_test,reg_pred)
print(score)
```

```python
Adjusted_R = 1 - (1 - score)*(len(y_test) - 1) / (len(y_test) - X_test.shape[1] - 1)
print(Adjusted_R)
```
